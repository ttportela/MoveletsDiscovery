```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

## Pre-Processing

```python
# PARAMS:
descFile = "descriptor.json"
inputFile = "foursquare.csv"

df = pd.read_csv(inputFile)
df.head(5)
```

```python
# SELECT FEATURES:
df = df[[
    'label',
    'tid',
    'time',
    'day',
    'poi',
    'price',
    'rating',
    'weather'
]]
df.head(5)
```

```python
X_train, X_test = train_test_split(df, test_size=0.33, random_state=1)
X_train.to_csv(r'train_'+inputFile)
X_test.to_csv(r'test_'+inputFile)
```
